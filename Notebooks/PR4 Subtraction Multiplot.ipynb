```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import healpy as hp
import os
```

```python

#getting beams
ellMax = 2500

F30,F44,F70,F100,F143,F217,F353,F545,F857 = rad = np.array([32.29,27.94,13.08,9.66,7.22,4.90,4.92,4.67,4.22])*(np.pi/10800)
SHbeam = []

for r in rad:
    B = []
    b = hp.gauss_beam(r,ellMax,pol=False)
    for l in range(0,ellMax+1):
        B.append(np.linspace(b[l],b[l],l+1))
    SHbeam.append(np.concatenate(B))

B30,B44,B70,B100,B143,B217,B353,B545,B857 = SHbeam
```

```python

#Loading and ordering PR4 FITS files
path = os.getcwd() + '/Desktop/CMB/PR4/Frequency Maps/Single-Frequency/Fits'
FITS = os.listdir(path)

def freq(string):
    return string[11:14]
FITS.sort(key=freq,reverse=True)
print(FITS)
```

```python

#Retrieving Dipole Data
pathd = os.getcwd() + '/Desktop/CMB/PR4/Dipole/Fits/dipole_CMB_n2048_K.fits'
fileDi = fits.open(pathd)
Dipole = np.concatenate(fileDi[1].data['T'])
```

```python

#defining 10 arcmin Beam
rad10 = 10*(np.pi/108000)
B = hp.gauss_beam(rad10,ellMax,pol=False)
b = []
for l in range(0,ellMax+1):
    b.append(np.linspace(B[l],B[l],l+1))
    
B10 = np.concatenate(b)
```

```python

Maps = []
Nside = []

for f in FITS:
    if f == '.DS_Store':
        continue
    
    mapName = f[11:14] + 'GHz'
    
    #Taking Data off FITS
    file = fits.open(os.path.join(path,f))
    Data = hp.reorder(np.concatenate(file[1].data['I_STOKES']),n2r=True)
    
    #Removing Dipole, 10 arcmin smoothing
    Data = hp.map2alm(Data,ellMax)
    Dipole = hp.map2alm(Dipole,ellMax)
    Data = hp.alm2map(Data,file[1].header['NSIDE'])
    Dipole = hp.alm2map(Dipole,file[1].header['NSIDE'])

    Data -= Dipole
    
    Data = hp.alm2map(hp.map2alm(Data,ellMax)*B10,file[1].header['Nside'])
    
    #adding to list
    Maps.append(mapName)
    Maps.append(Data)
    Nside.append(file[1].header['NSIDE'])

print(Maps)
```

```python

#pyramid plot

#subplots w/ mollview: https://stackoverflow.com/questions/37844221/healpy-plotting-how-do-i-make-a-figure-with-subplots-using-the-healpy-mollview
#increasing plot size: https://stackoverflow.com/questions/14770735/how-do-i-change-the-figure-size-with-subplots
fig, ax = plt.subplots(9,9,figsize=(20,20),dpi=150)

freq = ['857','545','353','217','143','100','070','044','030']
for c in range(1,9):
    for r in range(0,9):
        plt.axes(ax[r][0])
        hp.mollview(Maps[2*r+1],min=-0.01,max=0.01,norm='hist',flip='astro',cmap='Spectral_r',hold=True)
        plt.title(freq[r])
        if r >= c:
            plt.axes(ax[r][c])
            hp.mollview(hp.alm2map(hp.map2alm(Maps[2*r+1],ellMax)-hp.map2alm(Maps[2*r-2*c+1],ellMax)*SHbeam[8-r]/SHbeam[8-r+c],Nside[r]),min=-0.01,max=0.01,norm='hist',flip='astro',cmap='Spectral_r',hold=True)
            title = freq[r]+' - '+freq[r-c]+' * '+r'$\frac{'+freq[r]+' Beam}{'+freq[r-c]+' Beam}$'
            plt.title(title)
        if r < c:
            #removing empty axes: https://stackoverflow.com/questions/42038563/matplotlib-how-to-get-a-triangular-matrix-of-subplots
            ax[r][c].axis('off')

plt.subplots_adjust(wspace=0,hspace=0)
plt.savefig(os.getcwd() + '/Desktop/CMB/PR4/Adjusted Map Subtraction Pyramid.png')
plt.show()
```
